```python
import math

f_a = lambda x: math.sin(x) - 2 * math.pow(x, 2) + 0.5
f_b = lambda x: math.sqrt(1 - math.pow(x, 2)) - math.pow(math.e, x) + 0.1
f_c = lambda x:  math.pow(x,6) - 5 * x + 3 * x + 2
f_d = lambda x: math.log(1.5 * x) + 5 / (1 + math.pow(x, 2))
f_e = lambda x: math.pow(3, x) - 5 * x + 1
eps = 10**-4

MAXIMUM = -1
MINIMUM = 1
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
X = range(-20, 20)
Y = [f_a(x) for x in X]
ax.plot(X, Y)
plt.show()
```
